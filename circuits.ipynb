```python
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit import IBMQ, Aer, transpile, assemble
from qiskit.visualization import plot_histogram
from qiskit.circuit.library import MCXGate 

print("Imports Successful")
```

```python
scale=1  # scale factor for drawing the circuits

qA = QuantumRegister(3, name='q')
qout = QuantumRegister(1, name='out')


#c = ClassicalRegister(3, "bits")
circA= QuantumCircuit(qA, qout)

circA.x(qout)
circA.barrier()
circA.h(qout)
circA.h(qA)
circA.x([1,2])
gate = MCXGate(4-1)
circA.append(gate, list(range(4)))
circA.x([1,2])

circA.draw('mpl', scale=scale, plot_barriers=False)
```

```python
qB = QuantumRegister(3, name='q')
circB= QuantumCircuit(qB)
circB.h(qB)
circB.x([1,2])
circB.ccz(0,1,2)
circB.x([1,2])
circB.draw('mpl', scale=scale)
```

```python
qC = QuantumRegister(4, name='q')
ancilla = QuantumRegister(1, name='ancilla')

qout = QuantumRegister(1, name='out')


#c = ClassicalRegister(3, "bits")
circC= QuantumCircuit(qA, ancilla, qout)

gate = MCXGate(4-1)
circC.append(gate, list(range(3))+[4])
circC.barrier()

circC.ccx(0,1, 3)
circC.ccx(2,3, 4)
circC.ccx(0,1, 3)



circC.draw('mpl', scale=scale, plot_barriers=True)
```

```python
qD = QuantumRegister(3, name='q')
ancilla = QuantumRegister(1, name='ancilla')

qout = QuantumRegister(1, name='out')


#c = ClassicalRegister(3, "bits")
circD= QuantumCircuit(qD, ancilla, qout)

circD.h(qout)
gate = MCXGate(4-1)
circD.append(gate, list(range(3))+[4])
circD.h(qout)

circD.barrier()

circD.h(qout)
circD.ccx(0,1, 3)
circD.ccx(2,3, 4)
circD.ccx(0,1, 3)
circD.h(qout)



circD.draw('mpl', scale=scale, plot_barriers=True)
```

```python
qE = QuantumRegister(2, name='q')
qout = QuantumRegister(1, name='out')
c = ClassicalRegister(2, "bits")

circE= QuantumCircuit(qE, qout,c)
circE.x(qout)
circE.barrier()
circE.h(qE)
circE.h(qout)
circE.x(qE)
circE.ccx(0,1,2)
circE.x(qE)

circE.h(qE)
circE.x(qE)
circE.cz(0,1)
circE.x(qE)
circE.h(qE)
circE.measure(range(2),range(2))
circE.draw('mpl', scale=scale, plot_barriers=True)
```

```python
qF = QuantumRegister(3, name='q')
circF= QuantumCircuit(qB)
circF.h(qF)
circF.x([1,2])
circF.ccz(0,1,2)
circF.x([1,2])
circF.z(1)
print("Using the following, show that controls of Z can be any qubits.")
circF.draw('mpl', scale=scale)
```

```python
from qiskit import execute
simulator = Aer.get_backend('qasm_simulator')
job = execute(circE, simulator, shots=1024)

result = job.result()
counts = result.get_counts(circE)
display(plot_histogram(counts))
print(f"The answer is: {counts}")
```
